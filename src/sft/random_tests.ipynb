```python
import pandas as pd
from datasets import load_dataset

df = load_dataset("ronantakizawa/leetcode-assembly", split="train").to_pandas()

# Filter to one target toolchain for v1 and remove uneeded columns
df = df[df["architecture"] == "aarch64"]
df = df[df["compiler"] == "ARM64 gcc 15.2.0"]
df = df[[
    "problem_id",
    "problem_title",
    "difficulty",
    "c_source",
    "optimization",
    "assembly",
    
]].copy()


wide = (
    df.pivot_table(
        index=["problem_id", "problem_title", "difficulty", "c_source"],
        columns="optimization",
        values="assembly",
        aggfunc="first",
    )
    .rename(columns={"-O0": "assembly_00", "-O1": "assembly_01", "-O2": "assembly_02", "-O3": "assembly_03"})
    .reset_index()
)
wide.columns.name = None


wide
```

```python
# can we see the row 1 in more detail each column (not cutoff the text)
pd.set_option('display.max_colwidth', None)
wide.iloc[0]

# can i see this in a text file
wide.iloc[0].to_csv("wide.txt", index=False)
```
